```python
import os
from nltk.corpus import stopwords
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
import re

# Baca daftar stopword dari file
custom_stopword_file = "stopwords.txt"
with open(custom_stopword_file, 'r', encoding='utf-8') as stopword_file:
    custom_stopwords = stopword_file.read().splitlines()

# Inisialisasi stopword
stop_words = set(stopwords.words('indonesian'))

# Tambahkan stopwords khusus Anda ke dalam daftar stopwords
all_stopwords = set(stop_words).union(custom_stopwords)

# Inisialisasi stemmer Bahasa Indonesia
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# Fungsi untuk melakukan preprocessing pada satu file teks
def preprocess_file(file_path):
    tokens = []
    with open(file_path, 'r', encoding='utf-8') as file:
        # Baca baris pertama (tanpa stemming)
        first_line = file.readline().strip()

        # Baca sisa teks untuk proses stemming
        remaining_text = file.read()

    # Tokenisasi
    tokens = remaining_text.split()

    # Hapus token dari indeks 0 sampai 2
    tokens = tokens[3:]

    # Hapus stopword dan angka tunggal
    tokens = [word.lower() for word in tokens if word.lower() not in stop_words and len(word) > 1]

    # Stemming
    tokens = [stemmer.stem(word) for word in tokens]

    # Gabungkan kembali token menjadi teks
    preprocessed_text = first_line + '\n' + ' '.join(tokens)
    # print('hasil : ', preprocessed_text)


    output_path = file_path

    # Pastikan folder target sudah ada
    os.makedirs(os.path.dirname(output_path), exist_ok=True)


    # Tulis hasil preprocessing ke file target
    with open(output_path, 'w', encoding='utf-8') as file:
        file.write(preprocessed_text)
        print('Wrote to ' ,output_path, '\n')

# Fungsi untuk melakukan preprocessing pada semua file dalam suatu folder
def preprocess_folder(input_folder):
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            print(f'Processing {root}/{file}')
            file_path = os.path.join(root, file)
            preprocess_file(file_path)

# Path ke folder txts
folder_path = 'txts/'

# Lakukan preprocessing pada semua file dalam folder
preprocess_folder(folder_path)
```
